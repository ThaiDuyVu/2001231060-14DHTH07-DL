```python
# =========================
# BÀI 3 - CELL 1: 1D ARRAY
# So sánh 2 mảng 1 chiều
# =========================

import numpy as np

a = np.array([1, 5, 3, 8, 2])
b = np.array([2, 5, 1, 7, 4])

print("Mảng a:", a)
print("Mảng b:", b)

print("\n a > b:", np.greater(a, b))
print(" a >= b:", np.greater_equal(a, b))
print(" a < b:", np.less(a, b))
print(" a <= b:", np.less_equal(a, b))
```

```python
# =========================
# BÀI 7: MẢNG 10 PHẦN TỬ (15 → 55)
# IN TRỪ ĐẦU VÀ CUỐI
# =========================

import numpy as np

# Tạo mảng 10 phần tử trong khoảng [15, 55]
arr = np.random.randint(15, 56, size=10)

print("Mảng ban đầu:")
print(arr)

# Lấy các phần tử trừ đầu và cuối
result = arr[1:-1]

print("\nCác phần tử (trừ đầu và cuối):")
print(result)
```

```python
# =========================
# BÀI 8: ĐỔI DẤU PHẦN TỬ
# (9 ≤ x ≤ 15)
# =========================

import numpy as np

# Tạo mảng 20 phần tử trong khoảng [0, 20]
arr = np.random.randint(0, 21, size=20)

print("Mảng ban đầu:")
print(arr)

# Điều kiện các số trong khoảng [9, 15]
mask = (arr >= 9) & (arr <= 15)

# Đổi dấu các phần tử thỏa điều kiện
arr[mask] = -arr[mask]

print("\nMảng sau khi đổi dấu (9 → 15):")
print(arr)
```

```python
# =========================
# BÀI 10: MA TRẬN 10x10
# BIÊN = 1, BÊN TRONG = 0
# =========================

import numpy as np

# Tạo ma trận 10x10 toàn số 0
matrix = np.zeros((10, 10), dtype=int)

# Gán giá trị 1 cho các phần tử ở biên
matrix[0, :] = 1        # hàng đầu
matrix[-1, :] = 1       # hàng cuối
matrix[:, 0] = 1        # cột đầu
matrix[:, -1] = 1       # cột cuối

print("Ma trận 10x10:")
print(matrix)
```

```python
# =========================
# BÀI 14: MA TRẬN + VECTOR
# A (4x3) + y (3 phần tử)
# =========================

import numpy as np

# Tạo ma trận A kích thước 4x3
A = np.random.randint(1, 10, size=(4, 3))

# Tạo vector y gồm 3 phần tử
y = np.random.randint(1, 10, size=3)

print("Ma trận A:")
print(A)

print("\nVector y:")
print(y)

# Cộng vector y vào từng dòng của A (broadcasting)
result = A + y

print("\nKết quả A + y:")
print(result)
```

# =========================
# BÀI 19: TÌM PHẦN TỬ KHÔNG CÓ TRONG SERIES KHÁC
# =========================

```python
import pandas as pd

# Tạo Series sr1 và sr2
sr1 = pd.Series([1, 2, 3, 4, 5])
sr2 = pd.Series([2, 4, 6, 8, 10])

print("Series sr1:")
print(sr1)

print("\nSeries sr2:")
print(sr2)

# Tìm các phần tử trong sr1 mà không có trong sr2
# Sử dụng ~ (NOT) với isin()
items_not_in_sr2 = sr1[~sr1.isin(sr2)]

print("\nCác phần tử của sr1 không có trong sr2:")
print(items_not_in_sr2)
```

# BÀI 20: NUMPY UNION1D, INTERSECT1D TRÊN SERIES
# =========================

```python
import numpy as np
import pandas as pd

# Tạo 2 Series ngẫu nhiên với 20 phần tử
x = pd.Series(np.random.randint(0, 20, size=20))
y = pd.Series(np.random.randint(10, 30, size=20))

print("Series x:")
print(x)

print("\nSeries y:")
print(y)

# Tìm hợp (union) của hai Series
union_xy = np.union1d(x, y)

print("\nHợp của x và y (union1d):")
print(union_xy)

# Tìm giao (intersection) của hai Series
intersect_xy = np.intersect1d(x, y)

print("\nGiao của x và y (intersect1d):")
print(intersect_xy)
```

```python
import pandas as pd
import string

# Tạo Series x (số nguyên từ 0 đến 9)
x = pd.Series(range(10))

# Tạo Series y (chữ cái từ 'p' đến 'y')
y = pd.Series(list(string.ascii_lowercase[15:25])) # 'p' is the 16th letter (index 15), 'y' is the 25th letter

print("Series x:")
print(x)

print("\nSeries y:")
print(y)

# Thêm Series y vào Series x theo chiều dọc (vertical)
# Sử dụng pd.concat
vertical_concat = pd.concat([x, y])

print("\nKết quả nối dọc (vertical concat):")
print(vertical_concat)

# Thêm Series y vào Series x theo chiều ngang (horizontal)
# Sử dụng pd.concat với axis=1
horizontal_concat = pd.concat([x, y], axis=1)

print("\nKết quả nối ngang (horizontal concat):")
print(horizontal_concat)
```
